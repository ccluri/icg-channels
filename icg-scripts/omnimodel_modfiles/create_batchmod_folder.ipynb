```python
import os
from os import listdir
from os.path import isfile, isdir, join
from shutil import copyfile
```

```python
#load data dictionary from pkl file
iontype = "IH"
SM_NUM = 2
```

```python
### move all supermodel files to a new directory
# in order to batch run icg protocols

# note: folder needs to exist already (intentional)
path = "/media/icg-channels/icg-channels-"+iontype+"/"
newfolder = "/home/bill/Documents/ICG/data/icg-channels-"+iontype+"-sm"+str(SM_NUM)+"-accum"

flist = []
for (dirpath, dirnames, filenames) in os.walk(path):
    for file in filenames:
        if ("sm"+str(SM_NUM) in file and ".mod" in file):
            flist.append(dirpath+'/'+file)
print('Moving %d supermodel (%s SM%d) files to new directory.' %(len(flist),iontype,SM_NUM))

for f in flist:
    fsplit = newfolder + '/' + f.split("/")[-1]
    copyfile(f,fsplit)
```

```python
# move supermodels back to file structure

path = "/media/icg-channels/icg-channels-"+iontype+"/"
tracepath = "/home/bill/Documents/ICG/sm_traces/"+iontype+"/SM"+str(SM_NUM)+"/"

tracefiles = [f for f in listdir(tracepath) if isfile(join(tracepath, f))]
for f in tracefiles:
    fsplit = f.split("_",1)
    folder = fsplit[1]
    fsplit = folder.rsplit("_",1)
    folder = fsplit[0]+".mod/"
    if not os.path.exists(path+"/"+folder+"ICG_SM"+str(SM_NUM)+"/"):
        os.makedirs(path+"/"+folder+"ICG_SM"+str(SM_NUM)+"/")
    print("%s to %s" %(tracepath+f,path+"/"+folder+"ICG_SM"+str(SM_NUM)+"/"+f))
    copyfile(tracepath+f,path+"/"+folder+"ICG_SM"+str(SM_NUM)+"/"+f)
```
