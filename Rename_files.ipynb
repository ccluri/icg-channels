```python
import os
import re
import pickle
from glob import glob
from subprocess import call
from shutil import copyfile
```

```python
sub_channel = 'icg-channels-Ca'
all_channels = os.listdir(os.path.join('.', sub_channel))
mega_dict = {} #Default values dict of dict - to be pickled
```

```python
regex = re.compile(r'NEURON\s*\{(\s*[\w+,]\s*)*\s\}?', re.MULTILINE)
```

```python
with_suffix = 0
sans_suffix = 0
for ii, folder in enumerate(all_channels):
    if folder in ['.git', 'LICENSE', 'Readme.md']:
        pass
    else:
        with open(glob(os.path.join(sub_channel, folder, '*.mod'))[0]) as dummy_file:
            print('Checking files in ', folder, ii)
            mega_dict[folder] = {'orig_name': folder}
            txt_in_mod = dummy_file.read()
            try:
                nrn_str = regex.search(txt_in_mod).group()
                channel_name = re.search(r'\sSUFFIX\s*(.*)', nrn_str).group(1) #.group(1)                                                                                                                          
                mega_dict[folder]['new_name'] = channel_name.replace('\r','').strip()+'.mod'
                mega_dict[folder]['has_suffix'] = True
                with_suffix += 1
            except AttributeError:
                mega_dict[folder]['new_name'] = None
                mega_dict[folder]['has_suffix'] = False
                sans_suffix += 1
                print('No SUFFIX defined within NEURON in the mod file - not a distributed mechanism')
                pass
            
print('Total_files', len(mega_dict))
print('Files_with_suffix', with_suffix)
print('Files_without_suffix', sans_suffix)
```

```python
for ii, folder in enumerate(all_channels):
    if folder in ['.git', 'LICENSE', 'Readme.md']:
        pass
    else:
        try:
            print('Copying and renaming original file in', folder)
            os.chdir(sub_channel+'/'+ folder)
            if mega_dict[folder]['has_suffix']:
                copyfile(folder, folder+'.bac')
                os.rename(folder, mega_dict[folder]['new_name'])
            else:
                print('Not a distributed mechanism')
            call('nrnivmodl')
            os.chdir('../..')
        except FileNotFoundError:
            print('Perhaps already renamed! - check!')
            pass
            os.chdir('../..')
```

```python
with open(sub_channel+'.pkl', 'wb') as handle:
    pickle.dump(mega_dict, handle, protocol=2)
```
